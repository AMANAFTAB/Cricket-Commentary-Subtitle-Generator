```python
############################### THIS ONE WORKS #################################
    #####                                                              #####
        #####                                                 ######
             #####                                   ######
                  #####                         #####
                        #####         #####
                               #####
                        
### superimposing the commentary for the correspoding frame ###

###                                                                  ###
###           BUT THE COMMENTARY IS GOING BEYOND THE FRAME!          ###
###                                                                  ###


# Read the commentary from the text file and store it in a dictionary in the form (over_number: commentary)
commentary_file = r"C:\Users\amana\OneDrive\Desktop\cricket subtitle generation\miinn.txt"  # Replace with the actual commentary file path
commentary_dict = {}
with open(commentary_file, 'r') as file:
    for line in file:
        over=line.strip()
        commentary = next(file).strip()
        commentary_dict[over] = commentary


##### trying out for many frames #####
# setting a certain threshold to consider slighlty variable coordinates

def are_boxes_similar(box1, box2, threshold=20):
    for (x1, y1), (x2, y2) in zip(box1, box2):
        if abs(x1 - x2) > threshold or abs(y1 - y2) > threshold:
            return False
    return True

# Load bounding boxes from a file or define them here
# set of bounding box coordinates to compare with the actual bb_coordinates with a certain coordinates 
## **these coordinates are relative to the image(frame) dimensions**
### for example the coordinates given below are for images with dimensions(resolution) 960x544 
### but for an image with dimensions say 1920x1080 we will get different coordinates
### for the same frame14 with 960x544 ---> [[281, 475], [295, 475], [295, 499], [281, 499]]
###                          1920x1080 ---> [[562, 944], [590, 944], [590, 988], [562, 988]]
### so you can see that if the dimensions double, so does each point in the coordinate... 
bounding_boxes = [
    [[266, 472], [310, 472], [310, 498], [266, 498]],
    [[266, 472], [310, 472], [310, 500], [266, 500]],
    [[260, 470], [310, 472], [310, 498], [260, 495]],
    [[267, 471], [307, 472], [310, 498], [266, 498]],
    [[269, 474], [310, 472], [310, 499], [268, 501]]
]

text_position = (0, 420)  # Adjust the (x, y) coordinates as needed
font = cv2.FONT_HERSHEY_COMPLEX
font_scale = 0.56  # Adjust the font size as needed
text_color = (255, 255, 255)  # white color (BGR format)
background_color = (0, 0, 0)  # black background color (BGR format)
thickness = 1


# Path to folder containing frames
frames_folder = r"C:\Users\amana\OneDrive\Desktop\cricket subtitle generation\10 frames"
opdir = r"C:\Users\amana\OneDrive\Desktop\cricket subtitle generation\frames with subtitles" # output folder
# Initialize the main 'over_no' list
main_over_no = []

# Iterate through all frames in the folder
for frame_file in os.listdir(frames_folder):
    if frame_file.endswith(".jpg"):
        frame_path = os.path.join(frames_folder, frame_file)
        
        # Read the frame using OpenCV
        frame = cv2.imread(frame_path)
        op_filename = f"frame_with_commentary_{frame_file}"
        op_path = os.path.join(opdir, op_filename)
        
        # Perform OCR or other processing on the frame
        result = reader.readtext(frame)
        h = pd.DataFrame(result, columns=['bb', 'text', 'confidence'])
        column_values = [value for value in h['bb']]
        sb = []
        i = 0
        
        # Iterate through bounding boxes and column values
        while i < len(bounding_boxes):
            for x in column_values:
                if are_boxes_similar(x, bounding_boxes[i]):
                    sb.append(x)
                    break
            i += 1
            if len(sb) > 0:
                break
        
        over_no = []
        
        # Match bounding boxes and append text
        for bbox in sb:
            for index, row in h.iterrows():
                if row['bb'] == bbox:
                    over_no.append(row['text'])
                    break
        if len(over_no)==0:
            cv2.imwrite(op_path,frame)
            continue
        
        if over_no[0] in commentary_dict:
            commentary_text = commentary_dict[over_no[0]]
        else:
            cv2.imwrite(op_path, frame)
            continue
        
        # Get text size to determine background rectangle size
        text_size = cv2.getTextSize(commentary_text, font, font_scale, thickness)[0]
        background_size = (text_size[0] + 10, text_size[1] + 10)  # Adding some padding
        
        # Read the frame and add the commentary text
        frame = cv2.imread(frame_path)
        cv2.rectangle(frame, text_position, (text_position[0] + background_size[0], text_position[1] - background_size[1]), background_color, -1)
        
        cv2.putText(frame, commentary_text, text_position, font, font_scale, text_color, thickness)


        # Save frames with commentary to the output folder
        cv2.imwrite(op_path, frame)
        
        # Append the 'over_no' list of this frame to the main list
        main_over_no.extend(over_no)
```
